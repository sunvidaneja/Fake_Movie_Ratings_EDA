### Rotten Tomatoes

Let's first take a look at Rotten Tomatoes. RT has two sets of reviews, their critics reviews (ratings published by official critics) and user reviews. 

```python
plt.figure(figsize=(7,3), dpi=150)
sns.scatterplot(y='RottenTomatoes_User',x='RottenTomatoes', data=all_sites)
```

Let's quantify this difference by comparing the critics ratings and the RT User ratings. We will calculate this with RottenTomatoes-RottenTomatoes_User. Rotten_Diff here is Critics - User Score. So values closer to 0 means aggrement between Critics and Users. Larger positive values means critics rated much higher than users. Larger negative values means users rated much higher than critics.

```python
all_sites['RT_DIFF'] = (all_sites['RottenTomatoes']-all_sites['RottenTomatoes_User']).round(2)
```

Let's now compare the overall mean difference. Since we're dealing with differences that could be negative or positive, first take the absolute value of all the differences, then take the mean. This would report back on average to absolute difference between the critics rating versus the user rating.

**Calculate the Mean Absolute Difference between RT scores and RT User scores as described above.**

```python
all_sites['RT_DIFF'].abs().mean()
```

```python
plt.figure(figsize=(10,4),dpi=150)
sns.histplot(data=all_sites,x='RT_DIFF', bins=20, kde=True)
```

***absolute value* difference between Critics and Users on Rotten Tomatoes.**

```python
plt.figure(figsize=(10,4),dpi=150)
sns.histplot(data=all_sites,x=all_sites['RT_DIFF'].apply(abs), bins=20, kde=True)
```

**top 5 movies users rated higher than critics on average:**

```python
print('Users Love but Critics Hate')
all_sites.nsmallest(columns='RT_DIFF',n=5)[['FILM','RT_DIFF']]
```

**top 5 movies critics scores higher than users on average.**

```python
print('Critics love, but Users Hate')
all_sites.nlargest(columns='RT_DIFF',n=5)[['FILM','RT_DIFF']]
```

## Fandago Scores vs. All Sites

Finally let's begin to explore whether or not Fandango artificially displays higher ratings than warranted to boost ticket sales.

```python
scores_comparison = pd.merge(fandango,all_sites,on='FILM')
```

```python
scores_comparison.info()
```

```python
scores_comparison.head()
```

### Normalize columns to Fandango STARS and RATINGS 0-5 

```python
scores_comparison['RT_Norm'] = np.round(scores_comparison['RottenTomatoes']/20,1)
scores_comparison['RTU_Norm'] =  np.round(scores_comparison['RottenTomatoes_User']/20,1)
scores_comparison['Meta_Norm'] =  np.round(scores_comparison['Metacritic']/20,1)
scores_comparison['Meta_U_Norm'] =  np.round(scores_comparison['Metacritic_User']/2,1)
scores_comparison['IMDB_Norm'] = np.round(scores_comparison['IMDB']/2,1)
```

```python
scores_comparison.head()
```

**Creating norm_scores DataFrame that only contains the normalizes ratings.**

```python
norm_scores = scores_comparison[['STARS','RATING','RT_Norm','RTU_Norm','Meta_Norm','Meta_U_Norm','IMDB_Norm']]
```

```python
norm_scores.head()
```

### Comparing Distribution of Scores Across Sites


Now the moment of truth! Does Fandango display abnormally high ratings? We already know it pushs displayed RATING higher than STARS, but are the ratings themselves higher than average?


```python
fig = plt.figure(figsize=(15,6),dpi=150)
sns.kdeplot(data=norm_scores,clip=[0,5],shade=True,palette='Set1')
```

**Clearly Fandango has an uneven distribution. We can also see that RT critics have the most uniform distribution. Let's directly compare these two.** 


```python
fig = plt.figure(figsize=(15,6),dpi=150)
sns.kdeplot(data=norm_scores,x='RT_Norm',clip=[0,5],shade=True,palette='Set1')
sns.kdeplot(data=norm_scores,x='STARS',clip=[0,5],shade=True,palette='Set1')
```

```python
fig = plt.figure(figsize=(15,6),dpi=150)
sns.histplot(data=norm_scores,palette='Set1',bins=50)
```


### How are the worst movies rated across all platforms?

```python
fig = plt.figure(figsize=(15,6),dpi=150)
sns.clustermap(data=norm_scores,cmap='magma',col_cluster=False)
```

**Let's examine the top 10 worst movies. All Normalised scores**

```python
norm_films = scores_comparison[['STARS','RATING','RT_Norm','RTU_Norm','Meta_Norm','Meta_U_Norm','IMDB_Norm','FILM']]
```

```python
norm_films.nsmallest(10,'RT_Norm')
```

```python
plt.figure(figsize=(15,6),dpi=150)
worst_films = norm_films.nsmallest(10,'RT_Norm').drop('FILM',axis=1)
sns.kdeplot(data=worst_films,clip=[0,5],shade=True,palette='Set1')
plt.title("Ratings for RT Critic's 10 Worst Reviewed Films");
```
